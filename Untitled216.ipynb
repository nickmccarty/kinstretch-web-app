```python
import pandas as pd
from yt_dlp import YoutubeDL
from youtube_transcript_api import (
    YouTubeTranscriptApi,
    TranscriptsDisabled,
    NoTranscriptFound
)

QUERY = "kinstretch"
MAX_RESULTS = 100  # change as needed

ydl_opts = {
    "quiet": True,
    "extract_flat": True,
    "skip_download": True,
}

rows = []

with YoutubeDL(ydl_opts) as ydl:
    search_results = ydl.extract_info(
        f"ytsearch{MAX_RESULTS}:{QUERY}",
        download=False
    )

    for entry in search_results["entries"]:
        video_id = entry["id"]
        url = f"https://www.youtube.com/watch?v={video_id}"
        title = entry.get("title")
        creator = entry.get("channel")

        # --- Get transcript ---
        try:
            transcript_data = YouTubeTranscriptApi.get_transcript(video_id)
            transcript = " ".join(t["text"] for t in transcript_data)
        except (TranscriptsDisabled, NoTranscriptFound):
            transcript = None
        except Exception:
            transcript = None

        rows.append({
            "URL": url,
            "Title": title,
            "Creator": creator,
            "Transcript": transcript
        })

# --- Create DataFrame ---
df = pd.DataFrame(rows)

df
```

```python
import os
from yt_dlp import YoutubeDL

def download_video(url, out_dir="videos"):
    os.makedirs(out_dir, exist_ok=True)
    ydl_opts = {
        "outtmpl": os.path.join(out_dir, "%(id)s.%(ext)s"),
        "format": "mp4/bestvideo[ext=mp4]+bestaudio/best[ext=m4a]/best",
        "quiet": True,
    }
    with YoutubeDL(ydl_opts) as ydl:
        info = ydl.extract_info(url, download=True)
        filename = ydl.prepare_filename(info)
        # normalize extension to .mp4 if needed
        base, _ = os.path.splitext(filename)
        if not filename.endswith(".mp4") and os.path.exists(filename):
            new_name = base + ".mp4"
            os.rename(filename, new_name)
            filename = new_name
    return filename
```

```python
import cv2
import mediapipe as mp

mp_pose = mp.tasks.vision.PoseLandmarker
BaseOptions = mp.tasks.BaseOptions
PoseLandmarkerOptions = mp.tasks.vision.PoseLandmarkerOptions
VisionRunningMode = mp.tasks.vision.RunningMode

def extract_poses_from_video(video_path, model_path="pose_landmarker_full.task", frame_stride=5):
    # Download model once: https://ai.google.dev/edge/mediapipe/solutions/vision/pose_landmarker [web:15]
    options = PoseLandmarkerOptions(
        base_options=BaseOptions(model_asset_path=model_path),
        running_mode=VisionRunningMode.VIDEO,
    )
    poses = []
    cap = cv2.VideoCapture(video_path)
    fps = cap.get(cv2.CAP_PROP_FPS) or 30.0

    with mp_pose.create_from_options(options) as landmarker:
        frame_idx = 0
        while True:
            ret, frame = cap.read()
            if not ret:
                break
            if frame_idx % frame_stride != 0:
                frame_idx += 1
                continue

            rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            mp_image = mp.Image(image_format=mp.ImageFormat.SRGB, data=rgb)
            timestamp_ms = int(1000 * frame_idx / fps)
            result = landmarker.detect_for_video(mp_image, timestamp_ms)

            if result.pose_landmarks:
                # Take first person
                lm_list = result.pose_landmarks[0]
                wireframe = [
                    {
                        "x": lm.x,
                        "y": lm.y,
                        "z": lm.z,
                        "visibility": lm.visibility,
                    }
                    for lm in lm_list
                ]
                poses.append({
                    "timestamp_ms": timestamp_ms,
                    "landmarks": wireframe,
                })

            frame_idx += 1

    cap.release()
    return poses
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def plot_single_pose(landmarks, title="Pose Wireframe", figsize=(8, 10)):
    """Plot one frame's 33 landmarks with connections."""
    fig, ax = plt.subplots(1, 1, figsize=figsize)
    ax.set_xlim(-0.1, 1.1)
    ax.set_ylim(-0.1, 1.6)  # Flip Y for image coords
    ax.set_aspect('equal')
    ax.grid(True, alpha=0.3)

    # MediaPipe POSE_CONNECTIONS (key edges)
    POSE_CONNECTIONS = [
        (0,1),(0,2),(1,3),(2,4),(0,5),(0,6),(5,7),(7,9),(6,8),(8,10),
        (5,11),(6,12),(11,12),(11,13),(13,15),(12,14),(14,16),(9,11),(10,12)
    ]

    x_coords, y_coords = [], []
    for i, lm in enumerate(landmarks):
        x_coords.append(lm["x"])
        y_coords.append(lm["y"])
        color = 'red' if lm["visibility"] > 0.5 else 'lightgray'
        ax.plot(lm["x"], lm["y"], 'o', color=color, markersize=8)
        ax.text(lm["x"], lm["y"]-0.03, str(i), ha='center', fontsize=6)

    # Draw skeleton connections
    for start, end in POSE_CONNECTIONS:
        if landmarks[start]["visibility"] > 0.3 and landmarks[end]["visibility"] > 0.3:
            ax.plot([x_coords[start], x_coords[end]],
                   [y_coords[start], y_coords[end]], 'b-', alpha=0.6, lw=2)

    ax.invert_yaxis()
    ax.set_title(title)
    plt.tight_layout()
    plt.show()

# Example usage
if df["Poses"].iloc[0] and len(df["Poses"].iloc[0]) > 0:
    plot_single_pose(df["Poses"].iloc[0][0]["landmarks"], "Kinstretch Frame 1")
```

```python
from matplotlib.animation import FuncAnimation
import numpy as np

def animate_pose_sequence(poses, interval=200):
    """Animate pose sequence."""
    if not poses:
        print("No poses to animate")
        return

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

    def update(frame_idx):
        ax1.clear(); ax2.clear()
        ax1.set_xlim(-0.1, 1.1); ax1.set_ylim(-0.1, 1.6)
        ax2.set_xlim(-3, 3); ax2.set_ylim(-3, 3)
        ax1.grid(True, alpha=0.3); ax2.grid(True, alpha=0.3)

        landmarks = poses[frame_idx]["landmarks"]
        timestamp = poses[frame_idx]["timestamp_ms"] / 1000

        # 2D Image coords (left)
        plot_single_pose(landmarks, ax=ax1, title=f"Frame {frame_idx} ({timestamp:.1f}s)")

        # 3D World coords (right) - simplified
        x3d, y3d, z3d = [], [], []
        for lm in landmarks:
            x3d.append(lm["z"])  # Use Z as X for 3D view
            y3d.append(lm["x"])
            z3d.append(-lm["y"])

        ax2.scatter(x3d, y3d, c=z3d, cmap='viridis', s=50)
        ax2.set_title("3D World Coordinates")
        ax2.set_xlabel("Z (depth)"); ax2.set_ylabel("X")

    anim = FuncAnimation(fig, update, frames=len(poses), interval=interval, repeat=True)
    plt.tight_layout()
    plt.show()
    return anim

# Animate first video
anim = animate_pose_sequence(df["Poses"].iloc[0])
```

```python
def plot_joint_progression(poses, joint_indices=[11,12,23,24], colors=['r','g','b','o']):
    """Plot key joint trajectories over time."""
    fig, axes = plt.subplots(2, 2, figsize=(12, 8))
    axes = axes.ravel()

    for i, joint_idx in enumerate(joint_indices):
        times, xs, ys = [], [], []
        for pose in poses:
            lm = pose["landmarks"][joint_idx]
            times.append(pose["timestamp_ms"]/1000)
            xs.append(lm["x"])
            ys.append(lm["y"])

        axes[i].plot(times, xs, 'o-', label='X', alpha=0.7)
        axes[i].plot(times, ys, 's-', label='Y', alpha=0.7)
        axes[i].set_title(f"Joint {joint_idx}")
        axes[i].legend()

    plt.tight_layout()
    plt.show()

# Left/Right shoulders + hips
plot_joint_progression(df["Poses"].iloc[0], [11,12,23,24])
```
